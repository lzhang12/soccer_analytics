# Invasion Index

Invasion index : a measure of how close to the opponent’s goal a team plays during a match (i.e., its dangerousness) acceleration index: a measure of how fast a team reaches the closest position to the opponent’s goal.

```python
# add a tag_list column
tagfile = './data/Wyscout/tags2name.csv'
df_tag = pd.read_csv(tagfile, sep=';')
tagId2tag = dict(zip(df_tag['Tag'], df_tag['Description']))

match_events['tagId_list'] = match_events.tags.apply(lambda x: [i['id'] for i in x])
match_events['tag_list'] = match_events.tagId_list.apply(lambda x: [tagId2tag[i] for i in x])
match_events.head()
```

```python
# add player name column
playerfile = './data/Wyscout/players.json'
df_player = pd.read_json(playerfile)
playerId2player = dict(zip(df_player['wyId'], df_player['shortName']))
playerId2player[0] = 'None' # 

### This snippet shows events with playerId=0 and the followup event. 
### Interruption has playerId 0, and some free kicks missing player has playerId = 0.
# id_interruption = match_events[match_events.playerId==0].index.values
# id_followup = id_interruption + 1
# id_ = np.concatenate((id_interruption[:,np.newaxis], id_followup[:,np.newaxis]), axis=-1).reshape(-1)
# display(match_events.loc[id_])

match_events['player'] = match_events.playerId.apply(lambda x: playerId2player[x])
match_events.head()
```

```python
# add team name column
teamfile = './data/Wyscout/teams.json'
df_team = pd.read_json(teamfile)
teamId2team = dict(zip(df_team['wyId'], df_team['name']))

match_events['team'] = match_events.teamId.apply(lambda x: teamId2team[x])
match_events.head()
```
